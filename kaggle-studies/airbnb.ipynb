```python
airbnb = pd.read_csv("https://drive.google.com/u/0/uc?id=1SILFvyfG1IzM9OEh7MIX5v_HCNr3GpEw&export=download", low_memory=False)
airbnb.head(100)
```

airbnb["city"] = airbnb["city"].astype('category')
airbnb.dtypes

airbnb["city_cat"] = airbnb["city"].cat.codes
airbnb.head(30)

```python
# Data for Simple Linear Regression
from sklearn.model_selection import train_test_split

y = airbnb["price"]
X = airbnb.drop(["price"], axis=1)

X = X.drop(["city", "accommodates", "longitude", "latitude", "review_scores_rating", "number_of_reviews", "cleaning_fee", "bathrooms", "bedrooms", "beds", "property_type", "room_type", "availability_365", "host_identity_verified", "minimum_nights", "host_is_superhost", "cancellation_policy"], axis=1)

X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=42)

```

City_train = X_train.drop(["city", "longitude", "latitude", "review_scores_rating", "number_of_reviews", "security_deposit", "cleaning_fee", "accommodates", "bathrooms", "bedrooms", "beds", "property_type", "room_type", "availability_365", "host_identity_verified", "host_is_superhost", "cancellation_policy"], axis=1)

```python
from sklearn.linear_model import LinearRegression

# Simple Linear Regression to City
security_deposit_lr = LinearRegression()
security_deposit_lr.fit(X_train, y_train)
```

```python
# Visualising the Linear Regression results

# Visual libraries
import plotly.graph_objs as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=X_train["security_deposit"], y=y_train, name='security_deposit train', mode='markers'))
fig.add_trace(go.Scatter(x=X_train["security_deposit"], y=security_deposit_lr.predict(X_train), mode='lines', name='security_deposit regression'))

fig.update_layout(xaxis={"title": "Security Deposit"},
                  yaxis={"title": "PreÃ§o"})

fig.show()
```

```python
clf_score = security_deposit_lr.score(X_test,y_test)
pred_clf = security_deposit_lr.predict(X_test)
print("Regression Model Score is : ", round(clf_score*100))
```
